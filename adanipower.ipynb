```python
period=14
high,low,close= df['High'], df['Low'], df['Close']

plus_dm = high.diff()
minus_dm = low.diff()
plus_dm[plus_dm < 0] = 0
minus_dm[minus_dm > 0] = 0

hl = pd.DataFrame(high - low)
hc = pd.DataFrame(abs(high - close.shift(1)))
cl = pd.DataFrame(abs(low - close.shift(1)))
frames = [hl, hc, cl]
tr = pd.concat(frames, axis = 1, join = 'inner').max(axis = 1)
atr = tr.rolling(period).mean()

pdi = 100 * (plus_dm.ewm(alpha = 1/period).mean() / atr)
mdi = abs(100 * (minus_dm.ewm(alpha = 1/period).mean() / atr))
dx = (abs(pdi - mdi) / abs(pdi + mdi)) * 100
adx = ((dx.shift(1) * (period - 1)) + dx) / period
adx_smooth = adx.ewm(alpha = 1/period).mean()

df['+di'] = pd.DataFrame(pdi)
df['-di'] = pd.DataFrame(mdi)
df['adx'] = pd.DataFrame(adx_smooth)
```

```python
df
```

```python
ax = df.plot(y=["+di","-di","adx"],figsize=(20,4))
ax.axhline(25, color = 'grey', linewidth = 2, linestyle = '--')
```

```python
k = df['Close'].ewm(span=20, adjust=False, min_periods=20).mean()
d = df['Close'].ewm(span=40, adjust=False, min_periods=40).mean()
macd = k-d
df["MACD"] =macd
```

```python
plt.figure(figsize=(20,4))
plt.plot(df["MACD"])
```

```python
rate = 20
sma = df['Close'].rolling(rate).mean()
std = df['Close'].rolling(rate).std()
bollinger_up = sma + std * 2
bollinger_down = sma - std * 2
```

```python
plt.figure(figsize=(20,4))
plt.plot(bollinger_up,label='Bollinger Up', c='g')
plt.plot(bollinger_down,label='Bollinger down', c='r')
plt.plot(sma,label='SMA')
plt.legend()
```

```python
sensitivity = 150
t1 = ((macd-macd.shift(1))*sensitivity)
e1 = bollinger_up - bollinger_down
```

```python
trendUP = [0]*501
trendDown = [0]*501
for i in range(501):
    if t1[i] >=0 :
        trendUP[i] =t1[i]
        trendDown[i]=0
    else :
        trendUP[i]=0
        trendDown[i]=(-1*t1[i])
```

```python
colour1=['g']*501
colour2=['r']*501
for i in range(500):
    if trendUP[i+1] >=trendUP[i] :
        colour1[i+1]='g'
    else:
        colour1[i+1]='lime'
    if trendDown[i+1]>=trendDown[i] :
        colour2[i+1]='r'
    else:
        colour2[i+1]='orange'
```

```python
fig = plt.figure(figsize=(25,5))
plt.bar(df.index,trendUP,width=0.8,color=colour1,align='center')
plt.bar(df.index,trendDown,width=0.8,color=colour2,align='center')
plt.plot(e1,linewidth=2)
plt.axhline(25, color = 'grey', linewidth = 2, linestyle = '--')
```
